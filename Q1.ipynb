```python
import numpy as np
import pandas as pd
```

```python
rng=np.random.RandomState(1)
```

```python
X=np.dot(rng.rand(2,2),rng.randn(2,200)).T
X
```

```python
dataframe=pd.DataFrame(X,columns=['X','Y'])
dataframe
```

```python
import seaborn as sns
```

```python
sns.set()
```

```python
#plotting the data 
sns.lmplot(x='X',y='Y',data=dataframe,fit_reg=False)
```

```python
from sklearn.decomposition import PCA
```

```python
model=PCA(n_components=1)
model
```

```python
#fitting the model
model.fit(dataframe)
```

```python
#reducing the 2D data into 1D
dataframe_1D=model.transform(dataframe)
dataframe_1D
```

```python
#shape of the orginal data
dataframe.shape
```
